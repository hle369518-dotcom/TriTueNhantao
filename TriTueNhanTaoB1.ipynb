```python
#bai13
def tron_mang_cong(a, b):
    n = min(len(a), len(b))
    result = []

    # Cộng phần tử tương ứng
    for i in range(n):
        result.append(a[i] + b[i])

    # Thêm phần dư
    if len(a) > n:
        result.extend(a[n:])
    if len(b) > n:
        result.extend(b[n:])

    return result


# Test
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

print("Kết quả:", tron_mang_cong(a, b))
```

```python
#bai14

import random

# a. Tạo mảng
def tao_ma_tran(m, n):
    return [[random.randint(-10, 10) for _ in range(n)] for _ in range(m)]

# b. Xuất dòng k
def xuat_dong(a, k):
    print(a[k])

# c. Xuất cột k
def xuat_cot(a, k):
    for row in a:
        print(row[k])

# d. Tìm dòng có tổng lớn nhất
def dong_max(a):
    max_sum = sum(a[0])
    index = 0
    for i in range(len(a)):
        if sum(a[i]) > max_sum:
            max_sum = sum(a[i])
            index = i
    return index

# e. Cột có tích nhỏ nhất
def cot_tich_min(a):
    n = len(a[0])
    m = len(a)
    min_tich = None
    index = 0

    for j in range(n):
        tich = 1
        for i in range(m):
            tich *= a[i][j]
        if min_tich is None or tich < min_tich:
            min_tich = tich
            index = j
    return index

# f. Dòng chẵn, cột lẻ
def xuat_chan_le(a):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if i % 2 == 0 and j % 2 == 1:
                print(a[i][j], end=" ")
    print()

# g. TBC phần tử chẵn thuộc dòng lẻ
def tbc_chan_dong_le(a):
    tong = dem = 0
    for i in range(len(a)):
        if i % 2 == 1:
            for j in range(len(a[0])):
                if a[i][j] % 2 == 0:
                    tong += a[i][j]
                    dem += 1
    return tong / dem if dem != 0 else 0

# h. TBC phần tử biên
def tbc_bien(a):
    m, n = len(a), len(a[0])
    tong = dem = 0

    for i in range(m):
        for j in range(n):
            if i == 0 or i == m-1 or j == 0 or j == n-1:
                tong += a[i][j]
                dem += 1
    return tong / dem

# i. TBC tích phần tử không thuộc biên
def tbc_tich_noi(a):
    m, n = len(a), len(a[0])
    tich = 1
    dem = 0

    for i in range(1, m-1):
        for j in range(1, n-1):
            tich *= a[i][j]
            dem += 1

    return tich / dem if dem != 0 else 0


# Test
a = tao_ma_tran(4, 5)
for row in a:
    print(row)

print("Dòng max:", dong_max(a))
print("Cột tích min:", cot_tich_min(a))
print("TBC biên:", tbc_bien(a))
```

```python
import random

#bai15

class SinhVien:
    def __init__(self, ma, ten, namsinh, diem):
        self.ma = ma
        self.ten = ten
        self.namsinh = namsinh
        self.diem = diem

    # c. Đủ điều kiện lên lớp
    def dem_len_lop(ds):
        return sum(1 for sv in ds if sv.diem >= 5)

    # d. SV đủ 20 tuổi
    def sv_20_tuoi(ds, nam_hien_tai):
        return [sv for sv in ds if nam_hien_tai - sv.namsinh >= 20]

    # e. Đếm hệ DH
    def dem_he_dh(ds):
        return sum(1 for sv in ds if sv.ma[2:4] == "DH")

    # f. Tên "Lan"
    def dem_ten_lan(ds):
        return sum(1 for sv in ds if "Lan" in sv.ten)

    # g. Họ "Phan"
    def dem_ho_phan(ds):
        return sum(1 for sv in ds if sv.ten.startswith("Phan"))


# Test
ds = [
    SinhVien("02DH0001", "Phan Lan", 2003, 6.5),
    SinhVien("01CD0002", "Nguyen Lan", 2005, 4.0),
    SinhVien("02DH0003", "Phan An", 2002, 7.0)
]

print("Đủ lên lớp:", SinhVien.dem_len_lop(ds)) # Call static methods using class name
print("Hệ DH:", SinhVien.dem_he_dh(ds))
print("Tên Lan:", SinhVien.dem_ten_lan(ds))
print("Họ Phan:", SinhVien.dem_ho_phan(ds))

# Example for sv_20_tuoi, assuming current year is 2024
nam_hien_tai = 2024
sv_20 = SinhVien.sv_20_tuoi(ds, nam_hien_tai)
print("SV đủ 20 tuổi:", [sv.ten for sv in sv_20])
```

```python
#btvn1
def tron_mang_min(a, b):
    n = min(len(a), len(b))
    result = []

    for i in range(n):
        result.append(min(a[i], b[i]))

    result.extend(a[n:])
    result.extend(b[n:])

    return result
```

```python
#btvn2
#a. Tổng tam giác trên (chéo phụ)
def tong_cheo_phu_tren(a):
    n = len(a)
    tong = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:
                tong += a[i][j]
    return tong

# b. Đổi số âm -> trị tuyệt đối
def doi_am(a):
    return [[abs(x) for x in row] for row in a]

# c. Thay số chẵn = x
def thay_chan(a, x):
    return [[x if val % 2 == 0 else val for val in row] for row in a]

# d. Kiểm tra toàn chẵn
def toan_chan(a):
    return all(val % 2 == 0 for row in a for val in row)

# e. Kiểm tra đối xứng
def doi_xung(a):
    n = len(a)
    for i in range(n):
        for j in range(n):
            if a[i][j] != a[j][i]:
                return False
    return True
```
